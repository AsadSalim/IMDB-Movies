## One problem is that Pandas will read lists within the dataframe as strings, not as lists as shown below with the Stars column:

```python
for i, l in enumerate(df["Stars"]):
    print("list",i,"is",type(l))
```

```python
df["Stars"] = df["Stars"].apply(eval)
```

```python
for i, l in enumerate(df["Stars"]):
    print("list",i,"is",type(l))
```

```python
# After this step, our lists are finally recognized by Pandas as lists. Still, we can not 
# use the standard functions, because they are not made for whole lists
```

```python
# The idea is reducing the Series dimensions from 2 to 1, which would allow us to apply the typical 
# pandas functions again. For that, you can use this function:
def to_1D(series):
    return pd.Series([x for _list in series for x in _list])
```

```python
list_of_actors = to_1D(df["Stars"])
```

```python
list_of_actors = list_of_actors.apply(lambda x:x.strip())
```

```python
list_of_actors.value_counts()
```

```python
fig, ax = plt.subplots(figsize = (21,5))
ax.bar(list_of_actors.value_counts().index[:10],
        list_of_actors.value_counts().values[:10])
ax.set_ylabel("Frequency", size = 12)
ax.set_title("Top actors starring in movies in the list", size = 17)
plt.show()
```

### Robert De Niro and  then Tom Cruise are the most stars appearing in our movies list

```python
# For deeper analysis we can use the following tip
# Here we convert the list into a dataframe
Dataframe_of_stars = df["Stars"].apply(pd.Series)
```

## Now I will explore the relationship and correlation between the genres 

```python
def boolean_df(item_lists, unique_items):# Create empty dict
    bool_dict = {}
    
    # Loop through all the tags
    for i, item in enumerate(unique_items):
        
        # Apply boolean mask
        bool_dict[item] = item_lists.apply(lambda x: item in x)
            
    # Return the results as a dataframe
    return pd.DataFrame(bool_dict)
```

```python
genres_bool = boolean_df(df["Genres"], list_of_genres.value_counts().keys())
```

```python
genres_bool
```

```python
# From here, we can easily calculate association among these Genres and see which of them are closely related 
# One straight forward way is the Pearson correlation coefficient, which can also be used for binary data.
# Pandas has a built-int function for this.
genres_corr = genres_bool.corr(method = "pearson")
```

```python
genres_corr
```

```python
sns.heatmap(genres_corr, annot=False, cmap='RdBu', center=0, vmin=-1, vmax=1)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (20,15))
ax = sns.heatmap(genres_corr, annot=True, cmap='RdBu', center=0, vmin=-1, vmax=1, linewidth=1)
plt.show()
```

### The most correlated genres are Action and Adventure followed by Adventure and Animation

## Classify Movies based on Ratings [Excellent, Good, Average]

```python
def rating(ranking):
    if ranking >= 7:
        return 'Excellent'
    elif ranking >= 6:
        return 'Good'
    else:
        return 'Average'
```

```python
df['Rank_class'] = df['Rank'].apply(rating)
```
