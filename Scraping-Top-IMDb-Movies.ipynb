```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup as BS
from urllib.request import urlopen
%matplotlib inline
```

```python
# First we create and open a .csv file
# encoding is the name of the encoding used to decode or encode the file. This should only be used in text mode.
csv_file = open("IMDb-movies-data0.csv", mode = "w", encoding='utf-8')
header = "Movie_title,Year,Run_time,Genre,Rank,Metascore,Certificate,Director_Star,Gross_revenue\n"
csv_file.write(header)
```

```python
page_no = 10
for page in range(1, page_no+1):
    url = "https://www.imdb.com/list/ls006266261/?st_dt=&mode=detail&page="+ str(page) + "&sort=user_rating,desc"
    # Create a client-based request to get the URL
    client = urlopen(url)
    # Getting the HTML code for the full page
    html = client.read()
    # Closing the request
    client.close()
    # Creating an HTML parser using BeautifulSoup
    soup = BS(html, "html.parser")
    # Creating a container for the needed data
    containers = soup.find_all('div', {'class':'lister-item-content'})
    # Accessing the containers variable to extract the desired data from each container
    for container in containers:
        
        movie_title = container.a.get_text().replace(',', '.')
        
        Yr = container.findAll('span', {'class':'lister-item-year text-muted unbold'})
        year = Yr[0].text.strip()
        
        run_time = container.p.find('span', {'class':'runtime'}).text.strip()
        
        
        gnr = container.findAll('span', {'class':'genre'})
        genre = gnr[0].text.strip().replace(' ','').replace(',','-')
        
        Rnk = container.findAll('span', {'class':'ipl-rating-star__rating'})
        Rank = Rnk[0].text.strip()

        meta_score = container.find('span', {'class':'metascore'}).text if container.find('span', {'class':'metascore'}) else '-'

        certificate = container.find('span', {'class':'certificate'}).text if container.find('span', {'class':'certificate'}) else '-'

        Str = container.findAll('p', {'class':'text-muted text-small'})
        Dir_Star = Str[1].text.strip().replace('\n', '').replace('|', '').replace(',', '-').replace('   ','')

        
        Grs = container.find_all('span', {'name':'nv'})
        Gross = Grs[1].text if len(Grs) > 1 else '-'

        csv_file.write(movie_title + ',' + year + ',' + run_time + ',' + genre + ',' + Rank + ',' + meta_score + 
                       ',' + certificate + ',' + Dir_Star + ',' + Gross + '\n')

#         print(movie_title + ',' + year + ',' + run_time + ',' + genre + ',' + Rank + ',' + meta_score + ',' + certificate + ',' + Dir_Star + ',' + Gross)
#     print()
        
csv_file.close()
```
